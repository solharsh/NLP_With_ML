```python
train = pd.read_csv("headlines.csv")
```

```python
train.head()
```

```python
## We will create a classifier using embedding layer architecture
X=train['TITLE']
y=train['CATEGORY']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=200)
```

```python
train['TITLE'][5]
```

```python
enc=LabelEncoder()
```

```python
y_train=enc.fit_transform(y_train)
```

```python
enc.classes_
```

```python
y_train
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
seq_len=16
max_words=10000
```

```python
tokenizer=Tokenizer(num_words=max_words)
```

```python
### Split the text into words and assign an integer id
tokenizer.fit_on_texts(X_train.tolist())
```

```python
tokenizer.word_index
```

```python
## Create a sequence for each entry in the title column
sequence=tokenizer.texts_to_sequences(X_train.tolist())
```

```python
sequence
```

```python
## Pad the sequences
train_features=pad_sequences(sequence,maxlen=seq_len)
```

```python
train_features
```

```python
train_features.shape
```

```python
## Create test features
sequence=tokenizer.texts_to_sequences(X_test.tolist())
```

```python
sequence
```

```python
test_features=pad_sequences(sequence,maxlen=seq_len)
```

```python
test_features
```

```python
test_features.shape
```

```python
## Convert y_test and y_train to one hot encoded vector
from keras.utils import to_categorical
```

```python
y_train=to_categorical(y_train)
```

```python
y_train
```

```python
## Assemble the model
from keras.models import Sequential
from keras.layers import Dense, Embedding, Flatten
```

```python
model=Sequential()
model.add(Embedding(input_dim=max_words,output_dim=64,input_length=seq_len))
model.add(Flatten())
model.add(Dense(1024,activation='relu'))
model.add(Dense(4,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['acc'])
```

```python
model.fit(train_features,y_train,epochs=5,batch_size=32,validation_split=0.20)
```

```python
preds=model.predict_classes(test_features)
```

```python
preds
```

```python
pred_labels=enc.inverse_transform(preds)
```

```python
pred_labels
```
